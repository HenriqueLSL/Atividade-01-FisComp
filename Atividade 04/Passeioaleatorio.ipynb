# Passeio Aleat√≥rio 

```python
import numpy as np
import matplotlib.pyplot as plt

def passeio_aleatorio(N, p):
  """
  Simula um passeio aleat√≥rio unidimensional.

  """
  rng = np.random.default_rng()
  posicao = 0
  posicoes = [posicao]
  for _ in range(N):
    if rng.random() < p:
      posicao += 1
    else:
      posicao -= 1
    posicoes.append(posicao)
  return np.array(posicoes)


N = 1000
posicoes_p05 = passeio_aleatorio(N, 0.5)
posicoes_p08 = passeio_aleatorio(N, 0.8)

plt.plot(posicoes_p05, label='p = 0.5')
plt.plot(posicoes_p08, label='p = 0.8')
plt.legend()
plt.show()
```

```python
'''
Simula um passeio aleat√≥rio 100 vezes

'''

N = 1000
num_passeios = 100

plt.figure(figsize=(10, 6))

for _ in range(num_passeios):
  posicoes_p05 = passeio_aleatorio(N, 0.5)
  plt.plot(posicoes_p05, color='blue', alpha=0.1)

for _ in range(num_passeios):
  posicoes_p08 = passeio_aleatorio(N, 0.8)
  plt.plot(posicoes_p08, color='red', alpha=0.1)

plt.legend(['p=0.5 (azul)', 'p=0.8 (vermelho)'])
plt.show()
```
