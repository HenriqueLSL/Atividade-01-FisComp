```python
import numpy as np
import matplotlib.pyplot as plt

def gaussiana(x, amplitude, centro, largura):
  """
  Retorna os valores de uma função gaussiana.
  """
  return amplitude * np.exp(-((x - centro) / largura)**2)

# Define os parâmetros para as gaussianas
amplitudes = [1, 0.8, 0.6]
centro = [0, 2, 4]
largura = [1, 1.5, 0.8]

# Cria o vetor x
x = np.linspace(-5, 10, 500)  # Ajuste o intervalo conforme necessário


# Plota as funções gaussianas
plt.figure(figsize=(10, 6))  # Ajuste o tamanho da figura

for i in range(len(amplitudes)):
    y = gaussiana(x, amplitudes[i], centro[i], largura[i])
    plt.plot(x, y, label=f'Amplitude: {amplitudes[i]}, Centro: {centro[i]}, Largura: {largura[i]}')


plt.xlabel('x')
plt.ylabel('y')
plt.title('Funções Gaussianas')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def sinc(x):
  """
  Retorna os valores da função sinc(x) = sin(x)/x.
  """
  return np.sinc(x)

# Cria o vetor x
x = np.linspace(-10, 10, 500)

# Calcula os valores da função sinc(x)
y = sinc(x)

# Plota a função sinc(x)
plt.figure(figsize=(10, 6))
plt.plot(x, y)

plt.xlabel('x')
plt.ylabel('sinc(x)')
plt.title('Função sinc(x)')
plt.legend('s', loc='upper right')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
"""
 Retorna os valores da função exponencial y = exp(ax).
"""

# Define os valores de 'a'
a_valores = [-10, -1, 2, 10]

# Define o intervalo para x
x = np.linspace(-4, 4, 500)

# Plota as funções exponenciais para cada valor de 'a'
plt.figure(figsize=(10, 6))

for a in a_valores:
    y = np.exp(a * x)
    plt.plot(x, y, label=f'a = {a}')


plt.ylim(0, 10)
plt.xlabel('x')
plt.ylabel('exp(ax)')
plt.title('Função Exponencial y = exp(ax)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
"""
Retorna os valores da função y = ax^2 + bx + c.
"""

def plot_parabola(a, b, c):
  x = np.linspace(-10, 10, 400)
  y = a * x**2 + b * x + c

  delta = b**2 - 4 * a * c

  roots = []
  if delta > 0:
    roots = [(-b + np.sqrt(delta)) / (2 * a), (-b - np.sqrt(delta)) / (2 * a)]
  elif delta == 0:
    roots = [-b / (2 * a)]

  plt.figure(figsize=(8, 6))
  plt.plot(x, y)
  plt.xlabel("x")
  plt.ylabel("y")
  plt.title(f"Parábola: y = {a}x² + {b}x + {c} (Δ = {delta})")
  plt.grid(True)

  for root in roots:
    plt.plot(root, a * root**2 + b* root + c, 'ro', markersize=8)

  plt.show()


plot_parabola(1, -3, -4)
plot_parabola(1, -2, 1)
plot_parabola(1, 1, 1)
```
