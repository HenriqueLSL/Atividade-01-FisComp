# Monte Carlo 


```python
import numpy as np
'''
Calcula o valor de pi  através de uma aproximação de Monte Carlo.

'''
def pi_monte_carlo(n):
    num = np.random.uniform(0, 1, size=(n, 2))
    x = num[:, 0]
    y = num[:, 1]
    j = 0
    for i in range(0, n):
        r = (x[i]**2 + (y[i])**2)**0.5
        if r<= 1:
            j += 1
    return j, n
p = pi_monte_carlo(100000)[0]/pi_monte_carlo(100000)[1]
print(4*p)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def salvar_pontos(lista, pontos):
  lista += pontos

def pi_monte_carlo(n, retorna_pontos=False):
  salvar = lambda lista, pontos: None
  if retorna_pontos == True:
    salvar = salvar_pontos

  total = n
  num = np.random.uniform(0, 1, size=(total, 2))
  x = num[:, 0]
  y = num[:, 1]
  j = 0
  dentro = []
  fora = []
  for i in range(0, total):
      r = ((x[i])**2 + (y[i])**2)**0.5
      pontos = (x[i], y[i])
      
      if r <= 1:
          j += 1
          salvar(dentro, [pontos])
      else:
          salvar(fora, [pontos])

  pi_aproximado = 4 * (j / total)
  return pi_aproximado, dentro, fora
 
pi_aproximado, dentro, fora = pi_monte_carlo(10000, retorna_pontos=True)
print(pi_aproximado)
plt.scatter(*zip(*dentro), color='blue', s=1) 
plt.scatter(*zip(*fora), color='red', s=1)   
plt.show()
```
