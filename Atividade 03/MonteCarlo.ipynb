# Monte Carlo 


```python
import numpy as np
'''
Calcula o valor de pi  através de uma aproximação de Monte Carlo.

'''
def pi_monte_carlo(n):
    num = np.random.uniform(0, 1, size=(n, 2))
    x = num[:, 0]
    y = num[:, 1]
    j = 0
    for i in range(0, n):
        r = (x[i]**2 + (y[i])**2)**0.5
        if r<= 1:
            j += 1
    return j, n
p = pi_monte_carlo(100000)[0]/pi_monte_carlo(100000)[1]
print(4*p)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def pi_monte_carlo(n, retorna_pontos=False):
    '''
    Calcula o valor de pi através de uma aproximação de Monte Carlo.
    Separa os pontos dentro e fora do circulo de raio 1
    plota o gráfico dos pontos (azul = dentro) e (Vermelho = fora)

    '''
    total = n
    num = np.random.uniform(0, 1, size=(total, 2))
    x = num[:, 0]
    y = num[:, 1]
    j = 0
    dentro = []
    fora = []
    for i in range(0, total):
        r = ((x[i])**2 + (y[i])**2)**0.5
        if r <= 1:
            j += 1
            dentro.append((x[i], y[i]))
        else:
            fora.append((x[i], y[i]))
    p = j / total
    pi = 4 * p
    
    if retorna_pontos:
        return pi, dentro, fora
    else:
        return pi


pi, dentro, fora = pi_monte_carlo(10000, retorna_pontos=True)

x_dentro, y_dentro = zip(*dentro)
x_fora, y_fora = zip(*fora)

plt.figure(figsize=(6, 6))
plt.scatter(x_dentro, y_dentro, color='blue',  s=1)
plt.scatter(x_fora, y_fora, color='red',  s=1)


```
